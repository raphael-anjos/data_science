# Library

```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import plotly.express as px
from geopy.geocoders import Nominatim
```

# Functions

```python
#------------------------------------------
# Funções
#------------------------------------------

 
def show_dtypes (data):
    print(data.dtypes)
    return None

def show_dimensions (data):
    print ('Number of rows: {}'.format(data.shape[0]))
    print ('Number of columns: {}'.format(data.shape[1]))
    return None

def colect_geodata (data, cols):
    
    # Initialize aPI

    geolocator = Nominatim(user_agent = 'geopiExercise')

    # create empty

    data.loc[:, cols[0]] = 'NA'
    data.loc[:, cols[1]] = 'NA'

    for i in range(len(data)):
        print('Loop: {}/{}'.format(i, len(data)))

        query = str(data.loc[i, 'lat']) + ',' + str(data.loc[i, 'long'])
        response = geolocator.reverse(query)

        if cols[0] in response.raw['address']:
            data.loc[i, cols[0]] = response.raw['address'][cols[0]]

        if cols[1] in response.raw['address']:
            data.loc[i, cols[1]] = response.raw['address'][cols[1]]    
    return data        
```

# Extração

```python
# Loading DataSet
data = pd.read_csv('datasets/kc_house_data.csv')
```

```python
# data dimensions
show_dimensions(data)
```

```python
show_dtypes(data)
```

```python
colect_geodata(data,['house_number', 'road'])
```

# Transformação

```python
# Convert object to data

data['date'] = pd.to_datetime(data['date'])
```

```python
#descriptive statistics

num_atributes = data.select_dtypes(include=['int64', 'float64'])
pd.set_option('display.float_format', lambda x: '%.3f' % x)
# central tendency - media, mediana

media = pd.DataFrame (num_atributes.apply( np.mean, axis=0) )
mediana = pd.DataFrame (num_atributes.apply( np.median, axis=0) )

# dispersion - std, min, max
std = pd.DataFrame (num_atributes.apply( np.std, axis=0) )
min_ = pd.DataFrame (num_atributes.apply( np.min, axis=0) )
max_ = pd.DataFrame (num_atributes.apply( np.max, axis=0) )

df1 = pd.concat( [max_, min_, media, media, std], axis=1).reset_index( )
df1.columns = ['Atributes', 'Max', 'Mín', 'Media', 'Mediana', 'std']
```

```python
show_dimensions(df1)
```

```python
# Crie uma nova coluna chamada: "dormitory_type"
# Se o valor da coluna "bedrooms" for igual á 1 => 'studio'
# Se o valor da coluna "bedrooms" for igual á 2 => 'apartament'
# Se o valor da coluna "bedrooms" for maior que 2 => 'house'

data['dormitory_type'] = 'NA'

for i in range(len(data)):
    if data.loc[i,'bedrooms'] == 1:
        data.loc[i, 'dormitory_type'] = 'studio'
    elif data.loc[i, 'bedrooms'] == 2:
        data.loc[i, 'dormitory_type'] = 'apartament'
    else:
        data.loc[i, 'dormitory_type'] = 'house'
```

```python
show_dimensions(data)
```

```python
# Definir niveis de preço:
# 0 até 321.950              => Nivel 0
# Entre 321.950 até 450.000  => Nivel 1
# Entre 450.000 até 645.000  => Nivel 2
# Acima de 645.000           => Nivel 3

for i in range(len(data)):
    if data.loc[i, 'price'] <= 321950:
        data.loc[i, 'level'] = 0
    elif (data.loc[i, 'price'] > 321950) & (data.loc[i,'price'] <= 450000):
        data.loc[i, 'level'] = 1
    elif (data.loc[i, 'price'] < 450000) & (data.loc[i, 'price'] <= 645000):
        data.loc[i, 'level'] = 2
    else:
        data.loc[i, 'level'] = 3
```

```python
# map

houses = data[['id', 'lat', 'long', 'price', 'level']].copy() 

fig = px.scatter_mapbox(houses,
                 lat='lat',
                 lon='long',
                 color='level',
                 size='price',
                 color_continuous_scale=px.colors.cyclical.IceFire,
                 size_max=15,
                 zoom=10)

fig.update_layout(mapbox_style='open-street-map')
fig.update_layout(height=600, margin={'r':0,'l':0,'b':0,'t':0})
fig.show()
```

```python
def data_collect( path):
    #------------------------------------------
    # Extração
    #------------------------------------------
    # Loading DataSet
    data = pd.read_csv('datasets/kc_house_data.csv')

    # data dimensions
    show_dimensions(data)

    show_dtypes(data)
    return data

def data_transform (data):
    #------------------------------------------
    # Transformação
    #------------------------------------------
    # Loading DataSet
 
    # Convert object to data
    data['date'] = pd.to_datetime(data['date'])

    #descriptive statistics

    num_atributes = data.select_dtypes(include=['int64', 'float64'])
    pd.set_option('display.float_format', lambda x: '%.3f' % x)
    # central tendency - media, mediana

    media = pd.DataFrame (num_atributes.apply( np.mean, axis=0) )
    mediana = pd.DataFrame (num_atributes.apply( np.median, axis=0) )

    # dispersion - std, min, max
    std = pd.DataFrame (num_atributes.apply( np.std, axis=0) )
    min_ = pd.DataFrame (num_atributes.apply( np.min, axis=0) )
    max_ = pd.DataFrame (num_atributes.apply( np.max, axis=0) )

    df1 = pd.concat( [max_, min_, media, media, std], axis=1).reset_index( )
    df1.columns = ['Atributes', 'Max', 'Mín', 'Media', 'Mediana', 'std']

    show_dimensions(df1)

    # Crie uma nova coluna chamada: "dormitory_type"
    # Se o valor da coluna "bedrooms" for igual á 1 => 'studio'
    # Se o valor da coluna "bedrooms" for igual á 2 => 'apartament'
    # Se o valor da coluna "bedrooms" for maior que 2 => 'house'

    data['dormitory_type'] = 'NA'

    for i in range(len(data)):
        if data.loc[i,'bedrooms'] == 1:
            data.loc[i, 'dormitory_type'] = 'studio'
        elif data.loc[i, 'bedrooms'] == 2:
            data.loc[i, 'dormitory_type'] = 'apartament'
        else:
            data.loc[i, 'dormitory_type'] = 'house'

    show_dimensions(data)
    
    colect_geodata(data,['house_number', 'road'])

    # Definir niveis de preço:
    # 0 até 321.950              => Nivel 0
    # Entre 321.950 até 450.000  => Nivel 1
    # Entre 450.000 até 645.000  => Nivel 2
    # Acima de 645.000           => Nivel 3

    for i in range(len(data)):
        if data.loc[i, 'price'] <= 321950:
            data.loc[i, 'level'] = 0
        elif (data.loc[i, 'price'] > 321950) & (data.loc[i,'price'] <= 450000):
            data.loc[i, 'level'] = 1
        elif (data.loc[i, 'price'] < 450000) & (data.loc[i, 'price'] <= 645000):
            data.loc[i, 'level'] = 2
        else:
            data.loc[i, 'level'] = 3

    return data

def data_load(data):
    #------------------------------------------
    # Carga
    #------------------------------------------
    # Loading DataSet
 
    # map

    houses = data[['id', 'lat', 'long', 'price', 'level']].copy() 

    fig = px.scatter_mapbox(houses,
                     lat='lat',
                     lon='long',
                     color='level',
                     size='price',
                     color_continuous_scale=px.colors.cyclical.IceFire,
                     size_max=15,
                     zoom=10)

    fig.update_layout(mapbox_style='open-street-map')
    fig.update_layout(height=600, margin={'r':0,'l':0,'b':0,'t':0})
    fig.show()
    return None

if __name__ == '__main__':
    # ETL
    
    # Collect
    data_raw = data_collect ('kc_house_data.csv')
    # Transform
    data_processing = data_transform (data_raw)
    # Load
    data_load(data_processing)
```
